```python
from pandas import *
from matplotlib.pyplot import  *
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.tree import DecisionTreeRegressor
from sklearn import preprocessing
from sklearn.svm import SVC
```

```python
dataset = read_csv("proyecto_pre_procesado.csv",sep=";")
dataset
```

```python
x = dataset.iloc[:,:-1].values # DEPENDIENTES
#y = dataset.iloc[:,-1:].values # INDEPENDIENTES
y = dataset.iloc[:,-1:].values.flatten()
```

```python
#Variables de testeo y de entrenamiento
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
scaler = preprocessing.StandardScaler().fit(x)
```

```python
X_scaled = scaler.transform(x)
```

```python

# Ajuste de modelo con kernel gaussiano
model = SVC(kernel='rbf', C=500, gamma=1)
model.fit(x_train, y_train)
```

```python
model.score(x_train, y_train)
```

```python
predictions = model.predict(x_test)
print(predictions)
```
