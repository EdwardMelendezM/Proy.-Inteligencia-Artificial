```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from matplotlib import pyplot as plt
```

```python
dataset=pd.read_csv("proyecto_pre_procesado.csv",sep=";")
dataset
```

```python
x = dataset.iloc[:,:-1].values # DEPENDIENTES
#y = dataset.iloc[:,-1:].values # INDEPENDIENTES
y = dataset.iloc[:,-1:].values.flatten()
```

```python
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```

```python
model = SVC(kernel='linear',C=100)
model.fit(X_train, y_train)
```

```python
model.score(x,y)
```

```python
predictions = model.predict(x)
```

```python
suma=0
for valor in range(len(predictions)):
    porcentaje=dataset["Dureza"][valor] - predictions[valor] 
    if(porcentaje==0):
        suma+=1
    print(dataset["Dureza"][valor],predictions[valor],)
print("Porcentaje:",suma/len(predictions))
```
