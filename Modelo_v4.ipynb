```python
from pandas import *
from matplotlib.pyplot import  *
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.tree import DecisionTreeRegressor
from sklearn import preprocessing
from sklearn.svm import SVC
import numpy as np
from sklearn.svm import SVR
from sklearn.metrics import mean_squared_error
```

```python
dataset = read_csv("proyecto_pre_procesado.csv",sep=";")
dataset
```

```python
x = dataset.iloc[:,:-1].values # DEPENDIENTES
#y = dataset.iloc[:,-1:].values # INDEPENDIENTES
y = dataset.iloc[:,-1:].values.flatten()
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
```

```python
# -- Definir el modelo
model = SVR(kernel='rbf', C = 500, epsilon=0.1)
model.fit(x_train, y_train)
```

```python
model.score(x_test,y_test)
```

```python
predictions = model.predict(x_test)
print(predictions)
```

```python
suma=0
for valor in range(len(predictions)):
    porcentaje=y_test[valor] - predictions[valor] 
    if(-5<porcentaje<5):
        suma+=1
    print(y_test[valor],predictions[valor],porcentaje)
print("Porcentaje:",suma/len(predictions))
```
